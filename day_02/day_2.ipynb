```python
with open('./input.txt', 'r') as file:
    content = file.read()
```

```python
bag = {
    'red': 12,
    'green': 13,
    'blue': 14
}

impossible_games = []
for game in content.split('\n'):
    gameBag = game.split(': ')[1]
    set = [x.split(",") for x in gameBag.split(';')]
    
    # [['3 blue', ' 4 red'], ['1 red', ' 2 green', ' 6 blue'], ['2 green']]
    # check if number of balls is greater than number of balls in bag
    for stones in set:
        for stone in stones:
            stone = stone.strip().split(' ')
            if bag[stone[1]] < int(stone[0]):
                impossible_games.append(int(game.split(': ')[0].split(' ')[1]))
                break
        

# print(impossible_games) 
# sum from 1 to len of game and check if number is in impossible_games if so, don't add it to the sum
print(sum([x for x in range(1, len(content.split('\n'))+1) if x not in impossible_games]))
```

```python
# Part 2
# fewest number of cubes of each color that can be used to make the bag

powers = []
# loop over possible games
for i, game in enumerate(content.split('\n')):
    # if i+1 in impossible_games:
    #     continue

    gameBag = game.split(': ')[1]
    set = [x.split(",") for x in gameBag.split(';')]
    fewest_stones = { 'red': '', 'green': '', 'blue': ''}
    # [['3 blue', ' 4 red'], ['1 red', ' 2 green', ' 6 blue'], ['2 green']]
    # check if number of balls is greater than number of balls in bag
    for stones in set:
        for stone in stones:
            stone = stone.strip().split(' ')
            if fewest_stones[stone[1]] == '' or fewest_stones[stone[1]] < int(stone[0]):
                fewest_stones[stone[1]] = int(stone[0])
    print(fewest_stones)
    powers.append(fewest_stones['red'] * fewest_stones['green'] * fewest_stones['blue'])

print(sum(powers))
```
