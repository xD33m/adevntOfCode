```python
with open('./input.txt', 'r') as file:
    content = file.read()
    # print(content)
```

```python
# find first and last digit for each row
results = []
for row in content.split('\n'):
    firstDigit = ''
    lastDigit = ''
    for char in row:
        if char.isdigit():
            firstDigit = char
            break
    for char in reversed(row):
        if char.isdigit():
            lastDigit = char
            break
    results.append(firstDigit + lastDigit)

print(results)
    
    
```

```python
possibleSpelledNumbers = ['one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
mapWordToDigit = {
    'one': '1',
    'two': '2',
    'three': '3',
    'four': '4', 
    'five': '5',
    'six': '6',
    'seven': '7',
    'eight': '8', 
    'nine': '9',

}


def findFirstNum(str):
    smallestIndex = { 'num': '', 'index': len(str)}
    for num in possibleSpelledNumbers:
        if num in str:
            strIndex = str.find(num)
            if strIndex != -1 and strIndex < smallestIndex['index']:
                smallestIndex = { 'num': mapWordToDigit[num], 'index': strIndex }

    if smallestIndex['index'] == 0:
        return smallestIndex
    for i, char in enumerate(str):
        if char.isdigit() and i < smallestIndex['index']:
            smallestIndex = { 'num': char, 'index': i }
            break  
            
    return smallestIndex

def findLastNum(str):
    largestIndex = { 'num': '', 'index': 0}
    for num in possibleSpelledNumbers:
        if num in str:
            strIndex = str.rfind(num)
            if strIndex != -1 and strIndex > largestIndex['index']:
                largestIndex = { 'num': mapWordToDigit[num], 'index': strIndex }

    if largestIndex['index'] == len(str)-1:
        return largestIndex
    for i, char in enumerate(str):
        if char.isdigit() and i > largestIndex['index']:
            largestIndex = { 'num': char, 'index': i }
            
    return largestIndex

sum = 0
for row in content.split('\n'):
#    print(findFirstNum(row)['num'], findLastNum(row)['num'])
   sum += int(findFirstNum(row)['num']+findLastNum(row)['num'])

print(sum)
```
